# Leet Code problem # 27. Remove Element
# https://leetcode.com/problems/remove-element/description/


```python
nums = [3, 2, 2, 3]
val = 3

class Solution:
    def removeElement(self,nums,val):
        k = 0
        
        for i in range(len(nums)):
            if nums[i] != val:
                nums[k] = nums[i]
                k += 1
       
        return k
    
            
print(Solution().removeElement(nums, val))
```
